```python
import pandas as pd
from datetime import datetime
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_excel(r'C:\Users\agosa\OneDrive\Documents\GitHub\WorkSpace\taxpayer\data\DataNet.xlsx')
```

```python
df.columns
```

```python
X= df[['d1', 'd2', 'd3', 'd4', 'd5', 'd6']]
y= df[['predict']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X , y, random_state=104, test_size=0.25, shuffle=True)
```

```python
model = Sequential()
model.add(Dense(16, input_dim=6, activation='sigmoid'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
 
model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
 
model.fit(X_train, y_train, epochs=500)
 
scores = model.evaluate(X_test, y_test)
 
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
scores = model.evaluate(X_test, y_test)
 
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python

confusion_matrix(y_test,model.predict(X_test).round())
```
